## Outward protection rates of a new mask and a old masks are 0.7 and 0.5
## Inward protection rates of a new mask and a old masks are 0.6 and 0.4
## Mask production does not growth 

```python
from myfunctions import *
import numpy as np
import pandas as pd
from scipy.optimize import root
```

```python
def pmix_func_robust(x,m,I_0=0.05,vo=0.5):

    # 前兩期死亡人數
    d0 = (1-(17/18)**14) * 0.0138 * I_0
    Nation = Threeperiod_SIRD(S=1-I_0,I=I_0,q_0=m,q1_0=0,q1_1=0,σo=0.4, σn=0.6, δo=0.5,δn=0.7)
    Nation.serveralupdates(2)
    d1_phi, d1_n = Nation.d1_0, Nation.d1_1

    # t=1 q1_0 去登記拿到機率， q1_1 不去登記拿到機率，分配先給登記的有多才給不登記的
    q1_0_coef=x*(1-d0)*(1-m)
    q1_1_coef=(1-x)*(1-d0)*(1-m)+(1-d0)*m
    q1_0 = min(m/q1_0_coef,1)
    q1_1 = (m-q1_0_coef)/q1_1_coef if q1_0==1 else 0

    # t=2 的兩種情況
    if q1_0<1:
        # 上一期口罩不滿足所有登記的人，會有 1. 有登記沒拿到 q2_00 ，2.有登記有拿到 q2_1，3.沒登記沒拿到 q2_10
        q2_00_coef=(1-q1_0)*(1-d0)*x*(1-d1_phi)*(1-m)
        q2_10_coef=(1-d0)*( (1-x)*(1-d1_phi)*(1-m)+(1-d1_n)*m )
        q2_01_coef=q1_0*(1-d0)*x*(1-d1_phi)*(1-m)

        # 分配順序是 3→1→2
        q2_10 = min(m/q2_10_coef,1)
        q2_00 = min((m-q2_10_coef)/q2_00_coef,1) if q2_10==1 else 0
        q2_11 = 1 if q2_00==1 else 0
        q2_01 = (m-q2_10_coef-q2_00_coef)/q2_01_coef if q2_11==1 else 0

    else:
        # 口罩滿足所有有登記的人，會有 1.有登記有拿到 q2_1  ，2.沒登記沒拿到 q2_10 ，3.沒登記有拿到 q2_1
        q2_10_coef=(1-q1_1)*(1-d0)*( (1-x)*(1-d1_phi)*(1-m) + (1-d1_n)*m )
        q2_11_coef= q1_1 *(1-d0)*( (1-x)*(1-d1_phi)*(1-m) + (1-d1_n)*m )
        q2_01_coef= x*(1-d1_phi)*(1-d0)*(1-m)

        # 分配順序是 2→3→1
        q2_10 = min(m/q2_10_coef,1)
        q2_00 = 1 if q2_10==1 else 0
        q2_11 = min((m-q2_10_coef)/q2_11_coef,1) if q2_00==1 else 0
        q2_01 = (m-q2_10_coef-q2_11_coef)/q2_01_coef if q2_11==1 else 0

    phi_sign = q1_0*(1+q2_01+(1-q2_01)*vo) + (1-q1_0)*q2_00
    phi_nsign= q1_1*(1+q2_11+(1-q2_11)*vo) + (1-q1_1)*q2_10

    return phi_sign-phi_nsign

def Mix_computeProb_robust(x,y,m,I_0=0.05,vo=0.5):
    '''
        x 跟 y 分別代表 phi 跟 n 去登記的機率，然後要返回所有參數
    '''
    # 前兩期死亡人數
    d0 = (1-(17/18)**14) * 0.0138 * I_0
    Nation = Threeperiod_SIRD(S=1-I_0,I=I_0,q_0=m,q1_0=0,q1_1=0,σo=0.4, σn=0.6, δo=0.5,δn=0.7)
    Nation.serveralupdates(2)
    d1_phi, d1_n = Nation.d1_0, Nation.d1_1

    # t=1 q1_0 去登記拿到機率， q1_1 不去登記拿到機率，分配先給登記的有多才給不登記的
    q1_0_coef=x*(1-d0)*(1-m)+y*(1-d0)*m
    q1_1_coef=(1-x)*(1-d0)*(1-m) + (1-y)*(1-d0)*m
    q1_0 = min(m/q1_0_coef,1)
    q1_1 = (m-q1_0_coef)/q1_1_coef if q1_0==1 else 0

    # t=2 的兩種情況
    if q1_0<1:
        # 上一期口罩不滿足所有登記的人，會有 1. 有登記沒拿到 q2_00 ，2.有登記有拿到 q2_1，3.沒登記沒拿到 q2_10
        q2_00_coef=(1-q1_0)*(1-d0)*( x*(1-d1_phi)*(1-m)+y*(1-d1_n)*m )
        q2_10_coef=(1-d0)*( (1-x)*(1-d1_phi)*(1-m)+(1-y)*(1-d1_n)*m )
        q2_01_coef=q1_0*(1-d0)*( x*(1-d1_phi)*(1-m)+y*(1-d1_n)*m )

        # 分配順序是 3→1→2
        q2_10 = min(m/q2_10_coef,1)
        q2_00 = min((m-q2_10_coef)/q2_00_coef,1) if q2_10==1 else 0
        q2_11 = 1 if q2_00==1 else 0
        q2_01 = (m-q2_10_coef-q2_00_coef)/q2_01_coef if q2_11==1 else 0

    else:
        # 口罩滿足所有有登記的人，會有 1.有登記有拿到 q2_1  ，2.沒登記沒拿到 q2_10 ，3.沒登記有拿到 q2_1
        q2_10_coef=(1-q1_1)*(1-d0)*( (1-x)*(1-d1_phi)*(1-m)+(1-y)*(1-d1_n)*m )
        q2_11_coef= q1_1 *(1-d0)*( (1-x)*(1-d1_phi)*(1-m)+(1-y)*(1-d1_n)*m )
        q2_01_coef= (1-d0)*( x*(1-d1_phi)*(1-m)+y*(1-d1_n)*m )

        # 分配順序是 2→3→1
        q2_10 = min(m/q2_10_coef,1)
        q2_00 = 1 if q2_10==1 else 0
        q2_11 = min((m-q2_10_coef)/q2_11_coef,1) if q2_00==1 else 0
        q2_01 = (m-q2_10_coef-q2_11_coef)/q2_01_coef if q2_11==1 else 0
    
    return x,y,q1_0,q1_1,q2_10,q2_00,q2_11,q2_01,m
```

```python
I=0.01
m_vals=np.linspace(0.1,0.8,40)

# 解出 mix strategy 的均衡
result = map(lambda m: root(pmix_func_robust,1,args=(m,I),method='anderson',tol=10e-12),m_vals)
sol=list(result)
record=[(ele.x,ele.fun,ele.success) for ele in sol]
df=pd.DataFrame.from_records(record, columns =['x','fun','success'])
df['m_vals']=m_vals
# df

# 帶入數字把相對應的機率算出來
Xlist=df.x[:17]
Xlist=Xlist.append(pd.Series([1]*15+[0]*8),ignore_index=True)
Xlist

Ylist=[0]*40

param=map(lambda x,y,m: Mix_computeProb_robust(x,y,m,I_0=I,vo=0.5),Xlist,Ylist,m_vals)
paramdata= pd.DataFrame.from_records(list(param), columns =['p','q','q1_0','q1_1','q2_10','q2_00','q2_11','q2_01','m'])


# 把機率跟均衡策略帶進去算結果
result=map(lambda a,b,c,d,e,f,g,h,i: GRBT_evalDeath(p=a,q=b,q1_0=c,q1_1=d,q2_00=e,q2_10=f,q2_11=g,q2_01=h,q_0=i,S=1-I,I=I,t=300,σo=0.4,           σn=0.6, δo=0.5,δn=0.7),
           paramdata.p,paramdata.q,paramdata.q1_0,paramdata.q1_1,paramdata.q2_00,
           paramdata.q2_10,paramdata.q2_11,paramdata.q2_01,paramdata.m)
paramdata['D_vals']=list(result)

# 輸出表格
# paramdata.to_csv("data/0710 GRBT.csv",index=False)
paramdata
```

```python
I=0.01
m_vals=np.linspace(0.15,0.75,3)
result = root(pmix_func_robust,1,args=(0.15,I),method='anderson',tol=10e-12)

# 帶入數字把相對應的機率算出來
Xlist=result.x
Xlist=np.append(Xlist,[1,0])
Xlist

Ylist=[0]*3

param=map(lambda x,y,m: Mix_computeProb_robust(x,y,m,I_0=I,vo=0.5),Xlist,Ylist,m_vals)
paramdata= pd.DataFrame.from_records(list(param), columns =['p','q','q1_0','q1_1','q2_10','q2_00','q2_11','q2_01','m'])


# 把機率跟均衡策略帶進去算結果
result=map(lambda a,b,c,d,e,f,g,h,i: GRBT_evalDeath(p=a,q=b,q1_0=c,q1_1=d,q2_00=e,q2_10=f,q2_11=g,q2_01=h,q_0=i,S=1-I,I=I,t=300,σo=0.4, σn=0.6, δo=0.5,δn=0.7),
           paramdata.p,paramdata.q,paramdata.q1_0,paramdata.q1_1,paramdata.q2_00,
           paramdata.q2_10,paramdata.q2_11,paramdata.q2_01,paramdata.m)
paramdata['D_vals']=list(result)

# # 輸出表格
# paramdata.to_csv("data/0710 GRBT.csv",index=False)
paramdata
```
